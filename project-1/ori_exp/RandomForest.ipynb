```python
import pandas as pd
from sklearn.externals import joblib
from sklearn.preprocessing import LabelEncoder
import numpy as np

x = np.load("data.npy")
y = np.load("label.npy")

from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold

kf = StratifiedKFold(n_splits=3, random_state=None, shuffle=False)

avg_list = list()
idx = 0
for train_index, test_index in kf.split(x, y):
    print (idx)
    train = x[train_index, :]
    trainLabel = y[train_index]
    test = x[test_index, :]
    testLabel = y[test_index]
    
    clf = RandomForestClassifier(n_estimators=5, max_depth=2, random_state=0)
    
    clf.fit(train, trainLabel) 
    preTrainLabel = clf.predict(train)
    
    print ("Train")
    print (confusion_matrix(trainLabel, preTrainLabel))
    
    preTestLabel = clf.predict(test)
    
    print ("Test")
    print (confusion_matrix(testLabel, preTestLabel))
    
    testProb = clf.predict_proba(test)
    testProb = testProb[:, 1]
    
    fpr, tpr, thresholds = metrics.roc_curve(testLabel, testProb, pos_label=1)
    plt.figure()
    plt.plot(fpr, tpr)
    plt.show()
    print (roc_auc_score(testLabel, testProb))
    avg_list.append(roc_auc_score(testLabel, testProb))
    
    idx += 1
print (np.mean(avg_list))
```
